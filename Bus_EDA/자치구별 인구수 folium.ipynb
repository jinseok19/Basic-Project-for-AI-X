```python
import json
import pandas as pd
import folium
import requests


# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
people = pd.read_csv("1.3 seoul_people.csv", sep="\t", encoding = 'cp949')
people
```

```python
people = people.drop([0,1,2])
```

```python
people
```

```python
people = people[['자치구', '인구']]
people
```

```python
people = people.reset_index(drop= True)
```

```python
people_data = people['인구'].groupby(people['자치구']).mean()
print(people_data)
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m.choropleth(geo_data=seoul_geo,
             data=people_data, 
             fill_color='YlOrRd', # 색상 변경도 가능하다
             fill_opacity=0.5,
             line_opacity=0.2,
             key_on = 'properties.name',
             legend_name="지역구별 인구 수"
            )
m
```
