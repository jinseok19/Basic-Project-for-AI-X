```python
#2019.01_06중에 언제 지하철을 가장 많이 이용했을까?(기준:승하차총승객수)
import pandas as pd
import numpy as np
subway_raw = pd.read_csv('subway_raw.csv')
subway_raw
```

```python
subway_raw
```

```python
#월별로 승하차 승객수 더하고 그중 가장 높은 월 출력
subway_raw.info()
```

```python
#가설 : 월~6월중에 5월에 지하철 승객수가 많다? (기준 :승하자총승객수)
subway_raw.groupby('연월')\
                    .agg(sum_passenger = ('승하자총승객수','sum'))\
                    .sort_values('sum_passenger',ascending = False)
```

대립가설 : 5월은 다른 월별과 총 승객수의 차이가 있을 것이다
귀무가설 : 5월은 다른 월별과 총승객수의 차이가 없을것
이다. 
귀무가설이 틀림을 증명하면 대립가설이 맞다.

```python
#요일중에서 목요일에 지하철 승객수가 많다? (기준 :승하자총승객수)
subway_raw.groupby('요일')\
                    .agg(sum_passenger = ('승하자총승객수','sum'))\
                    .sort_values('sum_passenger',ascending = False)
```
