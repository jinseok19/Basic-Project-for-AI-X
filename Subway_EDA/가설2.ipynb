```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_style('whitegrid')
subway_raw = pd.read_csv('subway_raw.csv')
subway_raw
```

```python
#한글폰트깨짐 해결
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
font_list
plt.rcParams['font.family'] = 'HYGungSo-Bold'
```

```python
df_subway_01 = subway_raw.query('요일=="월"')
df_subway_02 = subway_raw.query('요일=="화"')
df_subway_03 = subway_raw.query('요일=="수"')
df_subway_04 = subway_raw.query('요일=="목"')
df_subway_05 = subway_raw.query('요일=="금"')
df_subway_06 = subway_raw.query('요일=="토"')
df_subway_07 = subway_raw.query('요일=="일"')
```

```python
df = pd.concat([df_subway_01,df_subway_02,df_subway_03,df_subway_04,df_subway_05,df_subway_06,df_subway_07])
df
```

```python
sns.boxplot(x='요일', y='승하자총승객수', data = df)
```

```python
# 유의미한 차이 확인-> One-Way ANOVA : Tukey's test
# reject이 유의미한 차이가 있으면 True 아니면 False
# p-value를 0.05 (PPT)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

posthoc = pairwise_tukeyhsd(df['승하자총승객수'], df['요일'], alpha=0.05)
print(posthoc)
fig = posthoc.plot_simultaneous()
```
