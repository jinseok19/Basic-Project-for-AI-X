```python
#1호선에서 가장 하차를 많이 하는 역은?
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib as plt
subway_raw = pd.read_csv('subway_raw.csv')
subway_raw
```

```python
#한글폰트깨짐 해결
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
font_list
plt.rcParams['font.family'] = 'HYGungSo-Bold'
```

```python
subway_raw
```

```python
df_subway = subway_raw.query('노선명 == "1호선"')\
                      .groupby('역명',as_index = False)\
                      .agg(n = ('하차총승객수','sum'))\
                      .sort_values('n',ascending = False)
df_subway
```

```python
sns.lineplot( data = df_subway, x = '역명', y = 'n')\
   .set(xlim = ('가산디지털단지','노량진'))
```
