```python
#2호선중에서 어느 역에서 승차가 가장 많이 발생할까? (Folium 역 표시)
import pandas as pd
import numpy as np
subway_raw = pd.read_csv('subway_raw.csv')
subway_raw
```

```python
df_subway = subway_raw.query('노선명 == "2호선"')\
                        .groupby('역명')\
                        .agg(n = ('승차총승객수','count'))\
                        .sort_values('n',ascending = False)
```

```python
geo_seoul = pd.read_csv('지하철노선위경도정보2.csv')
```

```python
geo_seoul.rename(columns = {'역이름':'역명'},inplace = True)
```

```python
new_geo_s = geo_seoul.query('호선 =="2호선"')
```

```python
#위도경도정보랑 승차총승객수 데이터 합치기
new_data = pd.merge(new_geo_s, df_subway, how = 'left', on = '역명')
index = len(new_data)
```

```python
import folium
#배경지도만들기
map= folium.Map(location=[37.56,127], #지도중심좌표
          zoom_start = 12,         #확대단계
          tiles = 'cartodbpositron') #지도종류

#시각화
for i in index:
    folium.CircleMarker([new_data['위도'][i],new_data['경도'][i]],radius=new_data['n'],popup=new_data['역명'][i],color='#3186cc',fill_color='#3186cc').add_to(map)
 
```
